```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['healthy','no_resp_illness_exposed','positive_asymp','positive_mild','positive_moderate','recovered_full','resp_illness_not_identified'],axis=1)
y = df.iloc[:,3:10]
```

```python
print(x.columns)
print(y.columns)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

Preprocessing Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x_train = pd.DataFrame(scaler.fit_transform(x_train))
x_test = pd.DataFrame(scaler.transform(x_test))
```

Building ANN Model

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
#from sklearn.pipeline import pipeline
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(126,input_dim=60,activation='relu'))
#classifier.dropout(0.2)

classifier.add(Dense(82,activation='relu'))
#classifier.dropout(0.2)

classifier.add(Dense(48,activation='relu'))
#classifier.dropout(0.2)

classifier.add(Dense(16,activation='relu'))
#classifier.dropout(0.2)

classifier.add(Dense(7,activation='softmax'))
```

```python
classifier.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
classifier.fit(x_train,y_train,epochs=100,batch_size=10)
```

```python
y_pred = classifier.predict(x_test)
```
